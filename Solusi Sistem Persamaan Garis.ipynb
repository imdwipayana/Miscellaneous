```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan garis pertama melewati titik (a,b) dan (c,d)
a=1 #Titik (a,b)
b=1
c=2 #Titik (c,d)
d=2
m1=(d-b)/(c-a)
print("Gradien garis pertama adalah {}".format(m1))
print("Persamaan garis pertama adalah y={}x+{} ".format(m1,-m1*a+b))
#y1 = m1*(x-a)+b

# Persamaan garis pertama kedua melewati titik (e,f) dan (g,h)
e=1 #Titik (e,f)
f=1
g=2  #Titik (g,h)
h=-2
m2=(h-f)/(g-e)
print("Gradien garis ke dua adalah {}".format(m2))
print("Persamaan garis kedua adalah y={}x+{} ".format(m2,-m2*e+f))
#y2=m2*(x-e)+f

if m1==m2:
    print("Gradiennya sama lho, jadi kemungkinannya ga ada solusi atau solusinya banyak.")
else:
    # Titik potong kedua garis pada (tp_x,tp_y)
    tp_x = (m1*a-b-m2*e+f)/(m1-m2)
    tp_y = m2*(tp_x-e)+f

    print("Titik potong untuk adalah ({},{})".format(tp_x,tp_y))

    x=np.arange(tp_x-5,tp_x+5,0.01)
    plt.figure(figsize=(4,4),dpi=500)
    
    plt.plot(x,m1*(x-a)+b,color='red',label='y={}x+{}'.format(m1,-m1*a+b))
    plt.plot(x,m2*(x-e)+f,color='blue',label='y={}x+{}'.format(m2,-m2*e+f))
    plt.plot(tp_x,tp_y,color='green',marker='o',markersize=4)
    plt.legend(loc='best')
    plt.show()
```
